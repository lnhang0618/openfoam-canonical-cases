```python
import pandas as pd

def analyze_csv(file_path):
    try:
        # 读取CSV文件
        df = pd.read_csv(file_path)

        # 1. 文件的基本结构
        print("文件的基本结构:")
        print(f"行数: {df.shape[0]}")
        print(f"列数: {df.shape[1]}")
        print()

        # 2. 列名（Header）
        print("列名（Header）:")
        print(df.columns.tolist())
        print()

        # 3. 数据类型
        print("每列的数据类型:")
        print(df.dtypes)
        print()

        # 4. 缺失值统计
        print("每列的缺失值统计:")
        print(df.isnull().sum())
        print()

        # 5. 数据样例（前5行）
        print("数据样例（前5行）:")
        print(df.head())

    except Exception as e:
        print(f"读取或分析CSV文件时出错: {e}")


# 使用示例
file_path = 'APG.csv'
analyze_csv(file_path)
```
